### functions

```python
def make_request(url):
    
    source = requests.get(url)
    
    if source.status_code == 200:
        print("Request Successful")
    elif source.status_code != 200:
        print("Request Unsuccessful")
        
    return source
```

```python
# get only text into source variable
# source = requests.get('https://stats.espncricinfo.com/ci/engine/records/averages/batting.html?class=2;current=2;id=6;type=team').text
source = requests.get('https://stats.espncricinfo.com/ci/engine/records/averages/batting.html?class=2;current=2;id=6;type=team').text
```

```python
# create beautiful soup object, 
# pass html we retreived from text 
# attribute and response object of html parser
soup = BeautifulSoup(source, 'lxml')
```

```python
print(soup.prettify())
```

```python
# find tbody tag in HTML
table_body = soup.find('th', class_="group-POWER-INDEX Table_TH") #text to only extract txt, no tags
```

```python
table_body[0:200]
```

```python
# remove line breaks from begining and end of string
table_body = table_body[2:-2]
```

```python
# seperate string into list of strings based on line break (\n\n\n)
table_list = table_body.split('\n\n\n')
```

```python
table_list[0:10]
```

```python
# loop list of strings, apply split method for each element in list
# split on line break, creating list of listsb
for index, element in enumerate(table_list):
    table_list[index] = element.split('\n')
```

```python
import pandas as pd
```

```python
# pass list of lists into pandas df
batting_avg_df = pd.DataFrame(data=table_list, columns=['Player',
                                                        'Span', 'Mat', 'Inns',
                                                       'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100',
                                                       '50', '0', '4s', '6s'])
```
