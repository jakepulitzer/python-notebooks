# Getting Reviews is a 2 step process
1. Go through amazon product catalog for "Clothing, Shoes, Jewelry and extract out matching products by their ASIN.
2. Go through the review data and parse out the matching reviews by ASIN.

The data used is 'per category' data for Clothing, Shoes, and Jewelry. 

THE EXAMPLE WAS NIKE, SO JUST CHANGE IT FOR MY BRAND


## 1. First, extract ASINS by brand

```
#first iterate through data and store it as a dict

#set counter to see how many products we have in the json
count = 0
#loading the json file (we already loaded it, but its not actually json yet
#initialize dictionary
allproducts = {}

#each line of data is a product and its metadata
for aline in loadedjson:
    
    #creating counter to know progress in processing the entire catalog
    count += 1
    if count % 100000 == 0:

        #only print count every 100k so we dont spam our output console
        print(count)

    #this data isnt true JSON, its actually python dictionaries printed as text. so, need to load a dictionary using the eval function
    #aline is a single product. So, aproduct will be a product

    aproduct = eval(aline)

    #makes a dict entry with the ASIN of the product as dict key and its metadata as nested dicts
    allproducts[aproduct['asin']] = aproduct
```

## 2. Next, explore product data to find commonalities.

* Product categories are "wishywashy" beacuse they can be categories (e.g. baby, house, home) or the actual brands! Remember, were already dealing with subset of product categories (shoes, clothing, jewelry), but we still need to find a lst of product id's for our specific brand.

*To do this, use the 'categories' metadata field to find the brand of interest!*

```
#get length of all products to track processing progress
len(allproducts)
```

```
#set count
count = 0
#initialize dict
allcategories = {}

for aproduct in allproducts:
    #set counter to know progress in processing entire catalog
    count += 1
    if count % 100000 == 0:
        #now we know there are 1.5 mil products, so build counter to tell us how processing is going.
        #when the counter reaches 1, we are done!
        print(count/1503384)

    #set up dict with just 1 prod to inspect and reference it later
    aproduct = allproducts[aproduct]

    #create dict entry for each product category
    #and count occurances of each category
    #2 for loops because categories is stored as lists of lists. 
    #acategory = string of each category
    if 'categories' in aproduct:
        for categories in aproduct['categories']:
            for acategory in categories:
                if acategory in allcategories:
                    allcategories[acategory] += 1
                if acategory not in allcategories:
                    allcategories[acategory] = 1
```

### Use code below to check and see which brands I might want to do!

```
#initiate list
sortedlist = []

#make tuple to put into sortedlist with the count for each category and the category name
for acategory in allcategories:
    sortedlist.append((allcategories[acategory], acategory))

#sort the tuples from highest to lowest. default is low to high so reverse=True
sortedlist = sorted(sortedlist, reverse=True)

#for every item in all of the sorted list, print it 
for item in range(0,100):
    print(sortedlist[item])
```

```
#Now, go through newly first dictionary and extract out the matching ASINs for the brand

##First, create a set where we will store ASINs
##We choose a set because we dont want duplicates

#change for my brand
allcarharttasins = set()
count = 0

for areview in allproducts:
    theproducts = allproducts[areview]
    count += 1
    if count % 100000 == 0:
        print(count/1503384)

    #iterate for each category for a product, again, any given product can be assigned multiple product categories
    for categories in theproducts['categories']:
        #each category is encoded as a list (even though they should be strings) so we need to iterate again
        for acategory in categories:
            #chcking to see if prod category matches my brand lowercasing category string incase capitalization might
            #get in the way of match
            if 'carhartt' in acategory.lower():
                #go ahead and store it to our set of Nike ASINs
                allcarharttasins.add(theproducts['asin'])

print(len(allcarharttasins))
```
