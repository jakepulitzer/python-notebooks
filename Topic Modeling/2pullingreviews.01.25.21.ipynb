```
import pickle
import json

##assign filename were trying to load in to a string variable

working_directory = '/content/drive/MyDrive/APRD6343'
working_file = '%s/reviews_Clothing_Shoes_and_Jewelry.json' % working_directory
loadedjson = open(working_file, 'r')
```

## Step 2 - parsing review data

```
##load review data into one big dictionary. 
##its the exact same process as loading the review data
count = 0
allreviews = {}
for aline in loadedjson:
    count += 1
    if count % 100000 == 0:
        print(count)
    areview = eval(aline)
    #arbitrarily using iteration count as key for dictionary
    #but dont have to do this
    allreviews[count] = areview

#check number of reviews
print(len(allreviews))
```

## Step 3 - Get reviews that match our brand

### Get asins for all carhartt products 

*(from other file)*

```
from time import sleep

#now go through reviews and pick out reviews that correspond to matching ASINS

allcarharttasins = []

for data in open('%s/allasins.txt' % working_directory, 'r'):
    #creates a comma for every time there is a split in the data, which is an asin
    asins = data.split(',')
    for anasin in asins:
        allcarharttasins.append(anasin)
print(allcarharttasins)
```

```
len(allcarharttasins)
```

### Match review asins with product asins

```
carharttreviews = {}
count = 0 
for areview in allreviews:
    count += 1
    if count % 100000 == 0:
        print(count/len(allreviews))
    #setting current review as a dict, so we can easily reference its entries
    thereview = allreviews[areview]
    theasin = thereview['asin']
    reviewerid = thereview['reviewerID']
    if theasin in allcarharttasins:
        #set key as something unique, if just did asin we would only have one review for each asin, 
        #with the last review the only one beign stored. Makes sure we dont just write it over
        thekey = '%s.%s' % (theasin, reviewerid)
        carharttreviews[thekey] = thereview

#all carhartt reviews are stored in carharttreviews!
print(len(carharttreviews))
```

## Save and write data

```
##save and write data as JSON dict file
json.dump(carharttreviews, open('%s/allcarharttreviews.json' % working_directory, 'w'))

##read data back in to make sure file isnt fucked up
allcarharttreviews = json.load(open('%s/allcarharttreviews.json' % working_directory, 'r'))
```
